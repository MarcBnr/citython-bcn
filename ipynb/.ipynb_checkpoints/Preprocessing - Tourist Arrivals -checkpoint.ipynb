```python
import pandas as pd
import numpy as np
```

```python
visitors = pd.read_csv('../data/gen_dataset.csv')
visitors.drop(columns=['Unnamed: 0'], inplace=True)
visitors.head()
```

```python
neighbourhoods = pd.read_csv('../data/neighbourhoods.csv')
neighbourhoods.head()
```

```python
ranks = []
for _, row in neighbourhoods.iterrows():
    ranks.append(float(row['RANK'].replace('%','')))

neighbourhoods['rank'] = ranks
```

```python
neighbourhoods.head()
```

```python
monthly_visitors = pd.read_csv('../data/monthly_visitors_percentage.csv')
monthly_visitors.head()
```

```python
visitors['neighbourhood_rank'] = neighbourhoods.iloc[visitors['N_Barri']]['rank'].tolist()
visitors['accomodation_rank'] = visitors['Type_of_Accomodation']
visitors['month_rank'] = monthly_visitors.iloc[visitors['Month']]['Percentage'].tolist()
visitors['1_over_days'] = 1 / visitors['Days']

visitors.drop(columns=['N_Barri','Days','Month','Type_of_Accomodation'], inplace=True)
```

```python
visitors
```

```python
visitors.to_csv('../data/gen_test_set.csv')
```
